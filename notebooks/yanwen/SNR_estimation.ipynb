```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy import ndimage
mpl.rc('image',interpolation = 'Nearest')
mpl.rc('image',origin = 'lower')
mpl.rc('image',cmap = 'jet')
mpl.rc('xtick',labelsize = 16)
mpl.rc('ytick',labelsize = 16)
mpl.rc('axes',labelsize = 16)
mpl.rc('legend',fontsize = 16)
mpl.rcParams['xtick.major.size'] = 8
mpl.rcParams['xtick.major.width'] = 3
mpl.rcParams['xtick.minor.size'] = 5
mpl.rcParams['xtick.minor.width'] = 1

mpl.rcParams['ytick.major.size'] = 8
mpl.rcParams['ytick.major.width'] = 3
mpl.rcParams['ytick.minor.size'] = 5
mpl.rcParams['ytick.minor.width'] = 1
mpl.rc('font',weight = 'bold')
mpl.rc('axes', linewidth= 3)
mpl.rc("axes", labelweight="bold")
```

```python
ratio = 2.1e-5/11000*2
```

```python
Nroi = 1e6
rep_rate = 120
SNR = 5.
beta = 0.1
ipm2 = np.logspace(0, 2, 100)*1e3
kavg = ipm2*ratio
pre_factor = 1
time = 2*(1+beta)*SNR**2/kavg**2/beta**2/Nroi/rep_rate/3600.
```

```python
plt.figure(figsize= (8,8))
plt.plot(ipm2, time, color = 'k', lw = 5, label = 'measurement time needed \n for one time point to get SNR = 5')
plt.xlabel('ipm2 (nJ)')
plt.ylabel('time (hours)')
plt.yscale('log')
plt.xscale('log')
plt.legend()
# plt.title('measurement time needed for a time point to get SNR = 5')
plt.tight_layout()
plt.grid('on', which = 'both')
```
